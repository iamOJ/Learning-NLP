# Plotting Word Frequency (Length)

```python
import nltk
from nltk.corpus import brown,inaugural,genesis
```

## Creating Frequency distribution and Cumulative Frequency Distribution
### The distribution would be created in the basis of the length of the words

```python
fd = nltk.FreqDist(len(word.lower()) 
                   for word in genesis.words('english-web.txt') 
                   if word.isalpha())
cfd = nltk.ConditionalFreqDist((len(word.lower()),word.lower()) 
                              for word in genesis.words('english-web.txt') 
                              if word.isalpha())
```

## Plot the Frequency Distribution for the words

```python
import matplotlib.pyplot as plt
fd.plot(figure=plt.figure(figsize=(8, 4)))
```

### What does the 'cfd' variable store?
It stores a dictionary of words as keys with their cumulative frequencies as values.

```python
cfd[15]
```

## Plot the Cumulative Frequency Distribution for the words

```python
cfd.plot(figure=plt.figure(figsize=(18, 16)))
```

```python
cfd.plot(conditions=range(20),figure=plt.figure(figsize=(18, 16)))
```
